B) Contar o número de películas de cada xénero.

```python
# mgenres=movies.aggregate([{'$unwind':'$genres'}])
df2 = pd.DataFrame(list(movies.aggregate([{'$unwind':'$genres'}])))
# df2 = pd.DataFrame(list(movies.find({''})))
pd.DataFrame(df2.groupby('genres').count()['_id'])
```

C) Contar cantas películas hai por ano.

```python
pd.DataFrame(df.groupby('year').count()['_id'])
```

D) Mirar se hai correlación entre imdb e rotten tomatoes.

```python
from pandas import json_normalize
df_tomatoes =json_normalize(df['tomatoes'])
df_imdb=json_normalize(df['imdb'])
df_imdb['rating'] = pd.to_numeric(df_imdb['rating'], errors='coerce')
df_imdb['votes'] = pd.to_numeric(df_imdb['votes'], errors='coerce')
df_tomatoes=pd.DataFrame(df_tomatoes['rotten'])
df_corr = pd.concat([df_tomatoes, df_imdb], axis=1)
df_corr.corr()

```

E) Contar cantas películas ten cada director

```python
df_direct = pd.DataFrame(list(movies.aggregate([{'$unwind':'$directors'}])))
pd.DataFrame(df_direct['directors'].value_counts())
```

F) Contar en cantas películas ten participado cada actor

```python
df_cast = pd.DataFrame(list(movies.aggregate([{'$unwind':'$cast'}])))
pd.DataFrame(df_cast['cast'].value_counts())
```

G) Contar cantas películas hai de cada idioma

```python
df_lang = pd.DataFrame(list(movies.aggregate([{'$unwind':'$languages'}])))
pd.DataFrame(df_lang['languages'].value_counts())
```

H) Cal é a media de premios das películas

```python
from pandas import json_normalize
df_awards =json_normalize(df['awards'])
media=df_awards['wins'].mean()
media
```

J) Amosar as películas que teñen alomenos 3 premios

```python

df_awards =json_normalize(df['awards'])
df_concat = pd.concat([df, df_awards], axis=1)
df_mas_3=df_concat[df_concat['wins'] >= 3]
df_mas_3[['title', 'wins']]
```
