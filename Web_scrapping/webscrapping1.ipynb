```python
import json
import requests

def chamadaepisodios(enderezo):
    solucion = []
    i = 1
    while True:
        response = requests.get(f"{enderezo}?page={i}")
        variables = json.loads(json.dumps(response.json()))
        for resultados in variables['results']:
            solucion.append(resultados['name'])
        if i >= variables["info"]["pages"]:
            break
        i = i+1
    solucion=set(solucion)#evita que se dupliquen datos
    solucion=list(solucion)
    return solucion

```

```python
episodios=chamadaepisodios(enderezo1)
episodios
```

```python
personaxes=chamadaepisodios(enderezo2)
personaxes
```

```python
import pandas as pd
dfP=pd.DataFrame(personaxes,columns=['nombres'])
dfP
```
